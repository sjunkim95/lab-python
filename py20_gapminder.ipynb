* GitHub에 저장된 [gapminder.tsv](https://github.com/JakeOh/202208_itw_java134_lab_python/raw/main/gapminder.tsv) 파일을 읽고 DataFrame을 생성
* 데이터 프레임 탐색
* (중복되지 않는) 국가 이름의 개수?
* (중복되지 않는) 대륙 이름?
* (중복되지 않는) 연도의 개수?
* 국가이름, 연도, 1인당 GDP 컬럼을 선택
* 1인당 GDP의 최댓값, 최솟값을 찾고, 그 row를 출력
* 기대수명(lifeExp)의 최솟값과 최댓값을 찾고, 그 행들을 출력
* 연도별 기대수명의 평균을 출력
* 연도별 1인당 GDP의 평균을 출력
* 연도별 대륙별 전세계 인구를 출력


```python
import numpy as np
import pandas as pd
```

```python
# GitHub에 저장된 파일 경로(이름)
file_path = 'https://github.com/JakeOh/202208_itw_java134_lab_python/raw/main/gapminder.tsv'
```

```python
# 파일을 읽고, DataFrame 생성
# CSV(Comma-Separated Values), TSV(Tab-Seperated Values)
gapminder = pd.read_csv(file_path, sep='\t')
```

```python
gapminder.tail()  # iloc[-5:]
```

```python
gapminder.info()  # 데이터프레임 정보: row/column index, non-null, dtype, ...
```

```python
gapminder.describe()  # 숫자 타입 변수들의 기술 통계량
```

```python
gapminder['country'].value_counts()  # 카테고리 타입의 기술 통계량 - 빈도수
```

```python
gapminder.continent.value_counts()
```

```python
# 중복되지 않은 국가 이름 개수
len(gapminder.country.unique())
```

```python
# 중복되지 않은 대륙 이름 개수
len(gapminder.continent.unique())
```

```python
# 중복되지 않은 연도 개수
len(gapminder.year.unique())
```

```python
# 컬럼 선택
gapminder[['country', 'year', 'gdpPercap']]
```

```python
# 1인당 GDP 최댓값
gapminder.gdpPercap.max()
```

```python
# 1인당 GDP 최솟값
gapminder.gdpPercap.min() # gapminder['gdpPercap'].min()
```

```python
gapminder[(gapminder.gdpPercap == gapminder.gdpPercap.max())|
          (gapminder.gdpPercap == gapminder.gdpPercap.min())]
```

```python
# 기대수명 최솟값, 최댓값
gapminder[(gapminder.lifeExp == gapminder.lifeExp.min()) |
          (gapminder.lifeExp == gapminder.lifeExp.max())]
```

```python
# 일본의 연도별 기대 수명
gapminder[gapminder.country == 'Japan'][['country', 'year', 'lifeExp']]
```

```python
# 르완다의 연도별 기대수명
gapminder[gapminder.country == 'Rwanda'][['country', 'year', 'lifeExp']]    
```

```python
# 연도별 기대수명 평균
gapminder.groupby('year').lifeExp.mean()
```

```python
# 연도별 1인당 GDP
gapminder.groupby('year').gdpPercap.mean()
```

```python
# 연도별 대륙별 인구수
gapminder.groupby(['year', 'continent'])['pop'].sum()
```
